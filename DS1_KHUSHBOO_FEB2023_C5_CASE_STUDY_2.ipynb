```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
import seaborn as sns 
from tabulate import tabulate               
import matplotlib as mat
```

```python
credit_data = pd.read_csv("DS1_C5_S4_Credit_Data_Hackathon.csv")
credit_data
```

```python
##removing "-" sign from the days_employed column as it is not necessary

credit_data['DAYS_EMPLOYED'] = credit_data['DAYS_EMPLOYED'].astype(str)
credit_data['DAYS_EMPLOYED'] = credit_data['DAYS_EMPLOYED'].str.replace("-",'')
credit_data
```

```python
credit_data.isnull().sum()
```

```python
##CHECKING THE MOST FREQUENT FOUND DATA IN THE MISSING VALUES COLUMNS

print(credit_data['NAME_TYPE_SUITE'].value_counts())
print("-"*70)
print(credit_data['OCCUPATION_TYPE'].value_counts())
print("-"*70)
print(credit_data['CNT_FAM_MEMBERS'].value_counts())
```

```python
##CALCULATING MEAN AND SETTING VALUES TO FILL THE MISSING VALUES
mean1 =  credit_data['AMT_GOODS_PRICE'].mean()
value1 = round(mean1,1)
print(value1)

print("-"*70)
mean2 =  credit_data['CNT_FAM_MEMBERS'].mean()
value2 = round(mean2)
print(value2)
```

```python
##FILLING THE MISSING VALUES
credit_data['AMT_GOODS_PRICE'] = credit_data['AMT_GOODS_PRICE'].fillna(value1)
credit_data['CNT_FAM_MEMBERS'] = credit_data['CNT_FAM_MEMBERS'].fillna(value2)

##replacing NAME_TYPE_SUITE & OCCUPATION_TYPE missing values by the most frequent values
credit_data['NAME_TYPE_SUITE'] = credit_data['NAME_TYPE_SUITE'].fillna('Unaccompanied')
credit_data['OCCUPATION_TYPE'] = credit_data['OCCUPATION_TYPE'].fillna('Laborers')
```

```python
##CHECKING IF THE VALUES GOT FILLED OR NOT
credit_data.isnull().sum()
```

```python
credit_data.nunique()
```

```python
def seprate_data_types(credit_data):
    categorical = []
    continuous = []
    for column in credit_data.columns:                
        if credit_data[column].nunique() < 19:       
                                                    
            categorical.append(column)
        else:
            continuous.append(column)
            
    return categorical, continuous 


categorical, continuous = seprate_data_types(credit_data)         


from tabulate import tabulate
table = [categorical, continuous]
print(tabulate({"Categorical":categorical,
                "continuous": continuous}, headers = ["categorical", "continuous"]))
```

```python
##defining a function for uniariate analysis of categorical datas.

def info_of_cat (col):
    print(f'Unique value in {col} are: {credit_data[col].unique()}')
    print("-"*60)
    print(f"Mode of {col} is {credit_data[col].mode()[0]}")
    print("-"*60)
    print(f"No of missing values in {col} is {credit_data[col].isnull().sum()}")
```
